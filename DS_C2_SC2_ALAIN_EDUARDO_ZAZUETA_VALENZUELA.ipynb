#c. Determinar la cantidad de vacunas aplicadas de cada compañía (en base a como lo reporta cada país en la columna vaccines)

```python
# NOTA: De acuerdo con las instrucciones se especifica utilizar SOLO el archivo "country_vaccinations.csv", con el cual no es posible 
# determinar cuantas vacunas se suministraron por compañia ya que solo se muestran los datos aglomerados de todas las compañias por
# país, el dato más cercano que se puede calcular es el siguiente:
print('Cantidad de vacunas por país y compañias: \n')
df_covid['total_vaccinations'].groupby([df_covid['country'], df_covid['vaccines']]).sum()
```

```python
# Sin embargo, dentro de la plataforma Kaggle en el mismo dataset existe un archivo extra llamado "country_vaccinations_by_manufacturer.csv" el cual SI
# categoriza las vacunas por país y por compañía, por lo que procedo a utilizarla para extraer la información solicitada:
df_covid_company = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/Practicas The Learning Gate/PANDAS Y NUMPY EN PYTHON/country_vaccinations_by_manufacturer.csv',  parse_dates=['date'])
df_covid_company.head(10)
```

#d. Obtener la cantidad de vacunas aplicadas en todo el mundo

```python
total_vacc_per_country_full=df_covid.groupby("country").total_vaccinations.tail(1)
print('Total de vacunas aplicadas en el mundo: {} vacunas'.format(total_vacc_per_country_full.sum()))
```

```python
total_vacc_per_country=df_covid.groupby("country").people_fully_vaccinated.tail(1)
print('Total de vacunas aplicadas en el mundo (Esquema completo): {} vacunas'.format(total_vacc_per_country.sum()))
```

#e. Calcular el promedio de vacunas aplicadas por país. 

```python
print('Promedio de aplicaciones de vacuna diaria por país: \n')
df_covid.groupby("country").daily_vaccinations.mean()
```

#f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo

```python
from pandas.core.indexes.datetimes import date

vacunas_29_01_21= df_covid[df_covid['date']=='2021-01-21']
print('Cantidad de vacunas aplicadas el día 29/01/2021: \t{}'.format(vacunas_29_01_21['total_vaccinations'].sum()))
```

#g. Crear un dataframe nuevo denoninado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccinations y daily_vaccinations_raw

```python
conDiferencias = df_covid
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias.tail(10)
```

#h. Obtener el período de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua

```python
print('Periodo de tiempo entre primera y última fecha de vacuna: {}'.format(df_covid['date'].max() - df_covid['date'].min()))
```

#i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter / )

```python
#El conteo se realizó de acuerdo a las instrucciones especificadas. Sin embargo, en realidad la columna 'vaccines'
#separa los nombres de vacunas por el caracter "," por lo que se realizó también la consulta con base a lo 
#anteriormente mencionado.
conCantidad = df_covid
conCantidad['canVac'] = conCantidad['vaccines'].str.split("/").str.len()
conCantidad.sample(n=8)
```

```python
conCantidad_opt = df_covid
conCantidad_opt['canVac'] = conCantidad['vaccines'].str.split(",").str.len()
conCantidad_opt.sample(n=8)
```

#j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = df_covid[df_covid['date'] < '2020-12-20']
antes20.sample(n=10)
```

#k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = df_covid[df_covid.vaccines.str.contains('Pfizer')]
pfizer.sample(n=10)
```

#l. Almacenar los dataframes generados (conDiferencias,conCantidad,antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter

```python
with pd.ExcelWriter('resultadosReto.xlsx',
                    date_format='DD-MM-YYYY',
                    datetime_format='DD-MM-YYYY HH:MM:SS',
                    mode='w'
                    ) as writer:
  conDiferencias.to_excel(writer, sheet_name='conDiferencias')
  conCantidad.to_excel(writer, sheet_name='conCantidad')
  antes20.to_excel(writer, sheet_name='antes20')
  pfizer.to_excel(writer, sheet_name='Pfizer')
```
